```python
df_columns=df.columns
df_columns
```

```python
for i in df_columns:
    df= df.rename(columns={i:i.replace(" ","_")}) 
df
```

```python
nan_list=[]
for column_name in df.columns:
    if df[column_name].isnull().any():
        print(f"{column_name} True")
        nan_list.append(f"{column_name}")
    else:
        print(f"{column_name} False")
```

```python
nan_list
```

```python
for i in nan_list:
    df[i]=df[i].replace("nan",None)
```

```python
dropColumn=["Latitude","Longitude","Temperature(F)"]
for i in dropColumn:
    df=df.drop(columns=[i])
```

```python
df
```

```python
i='Open/Closed_Space'
df=df.rename(columns={i:i.replace("/","_")}) 
```

```python
j='Influencing_Event/Event'
df=df.rename(columns={j:j.replace("/","_")}) 
```

```python
k='No._of_Suicide_Blasts'
df=df.rename(columns={k:k.replace(".","")}) 
```

```python
l='Explosive_Weight_(max'
df=df.rename(columns={l:l.replace("(","")}) 
```

```python
k='Temperature_C)'
df=df.rename(columns={k:k.replace(")","")}) 
```
